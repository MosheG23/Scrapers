## Creating data average per season

```python
first82Avg = pd.DataFrame()
# stats.insert(0, 'Year', year)
# NBA season we will be analyzing
for year in range(1975,2020):

    # URL page we will scraping (see image above)
#     url = ***url***

    html = urlopen(url)

    soup = BeautifulSoup(html)
    #     print(soup)
        # use findALL() to get the column headers
    soup.findAll('tr', limit=2)

        # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    col = ['Name of Song', 'Singer', 'Year']

        # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]

    # getting ID


    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    stats = pd.DataFrame(player_stats, columns = headers)
    links = [[a for a in rows[i].findAll('a')]
            for i in range(len(rows))]
    playersHref = '(?:href=")([^"]+)'
    # linksCopy = (str)(links)
    ids = re.findall(playersHref, (str)(links))
    idsList = []
    for i in range(0, len(ids) - 1):
        if('players' in ids[i]):
            currID = ids[i].split('/')[3]
            idsList.append(currID.split('.')[0])
    stats = stats.assign(Year = year)
    stats.Player = stats.Player.astype(str)
    stats = stats[stats['Player']!='None']
    stats['ID'] = idsList
    stats = stats.drop_duplicates(subset = 'ID', keep='first')
    # stats = stats[]

    first82Avg = pd.concat([first82Avg, stats])
first82Avg = first82Avg.reindex(columns=['ID', 'Year'] +list(first82Avg.columns[:-2]))
```

### Creating player career total

1. Creating the columns for the table
+ Creating players list (by 82 games dataframe because there are players that played a season game that didn't played a playoff game)
+ Going through both dataframes (82 games & playoff) by player and adding them for one dataframe

```python
columnsTable = first82Tot.columns
coulmns = (columnsTable.to_series())
columnsTable
```

```python
first82TotalPlayerListID = first82Tot.drop_duplicates(subset = 'ID', keep='first')
playerIDList = first82TotalPlayerListID.iloc[:, 0].tolist()
playerIDListSize = len(playerIDList)
```

```python
newTableCoulmns = ['ID', 'Player', 'Years', 'Pos', 'G', 'GS',  'MP', 'FG', 'FGA',
       'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA',
       'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'PIR']
newTableCoulmns = pd.Series((v for v in newTableCoulmns))
```

```python
careerTot = pd.DataFrame()
for y in range(0, playerIDListSize):
    df = pd.DataFrame()
    currID = playerIDList[y]
    curr82Player = first82Tot[first82Tot['ID'] == currID]
    currPlayoffPlayer = playoffTot[playoffTot['ID'] == currID]
    avg = []
#     player ID
    avg.append(currID)
#     Player Name
    avg.append(curr82Player['Player'].values[0])
#     Num of Years
    numOfYears = (int)(curr82Player.describe()['Age'][0])
    avg.append(curr82Player.describe()['Age'][0])
#     Position
    avg.append(curr82Player.describe(include='all')['Pos'][2])
#     Total Games
    totalGames = curr82Player['G'].sum() + currPlayoffPlayer['G'].sum()
    avg.append(totalGames)
    for x in range(7, coulmns.size):
        currCol = (str)(columnsTable[x])
        stats = round((curr82Player[currCol].sum() + currPlayoffPlayer[currCol].sum()), 2)
        avg.append(stats)
    avg = pd.DataFrame(avg)
    avg = pd.DataFrame(avg.transpose())
    avg.columns = newTableCoulmns
    avg['FG%'] = (avg['FG'] / avg['FGA']) * 100
    avg['3P%'] = (avg['3P'] / avg['3PA']) * 100
    avg['2P%'] = (avg['2P'] / avg['2PA']) * 100
    avg['eFG%'] = ((avg['FG'] + 0.5 * avg['3P']) / avg['FGA']) * 100
    avg['FT%'] = (avg['FT'] / avg['FTA']) * 100
    careerTot = pd.concat([careerTot,avg])
```

## Creating Career avarge

```python
careerAVG = pd.DataFrame()
for y in range(0, playerIDListSize):
    df = pd.DataFrame()
    currID = playerIDList[y]
    curr82Player = first82Tot[first82Tot['ID'] == currID]
    currPlayoffPlayer = playoffTot[playoffTot['ID'] == currID]
    avg = []
#     player ID
    avg.append(currID)
#     Player Name
    avg.append(curr82Player['Player'].values[0])
#     Num of Years
    numOfYears = (int)(curr82Player.describe()['Age'][0])
    avg.append(curr82Player.describe()['Age'][0])
#     Position
    avg.append(curr82Player.describe(include='all')['Pos'][2])
#     Total Games
    totalGames = curr82Player['G'].sum() + currPlayoffPlayer['G'].sum()
    avg.append(totalGames)
#     Total Game Started
    totalGamesS = curr82Player['GS'].sum() + currPlayoffPlayer['GS'].sum()
    avg.append(totalGamesS)
    for x in range(8, coulmns.size):
        currCol = (str)(columnsTable[x])
        stats = round((curr82Player[currCol].sum() + currPlayoffPlayer[currCol].sum()) / totalGames, 3)
        avg.append(stats)
    avg = pd.DataFrame(avg)
    avg = pd.DataFrame(avg.transpose())
    avg.columns = newTableCoulmns
    avg['FG%'] = (avg['FG'] / avg['FGA']) * 100
    avg['3P%'] = (avg['3P'] / avg['3PA']) * 100
    avg['2P%'] = (avg['2P'] / avg['2PA']) * 100
    avg['eFG%'] = ((avg['FG'] + 0.5 * avg['3P']) / avg['FGA']) * 100
    avg['FT%'] = (avg['FT'] / avg['FTA']) * 100
    careerAVG = pd.concat([careerAVG,avg])
careerAVG
```
